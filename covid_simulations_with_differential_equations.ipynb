```julia
using Interact
using Plots
using DifferentialEquations
using ParameterizedFunctions
```

```julia
println("Parameters:")
println("recovery rate - everyday, this fraction of infected people recovers from disease")
println("contagion rate - everyday, everyone spreads disease to this many people")
println("infected at start - fraction of population infected at start of simulation")
println("end time - last day to be displayed on graph")
@manipulate for k in slider(0.0:0.01:0.5, label="recovery rate"),
                b in slider(0.0:0.01:1, label="contagion rate"), 
                i0 in slider(1.27e-6:1.27e-6:0.01, label="infected at start"), 
                tend in slider(100.0:10.0:300.0, label="end time")
    function covid!(du,u,p,t)
        b, k = p
        du[1] = (-b) * u[1] * u[2]                 
        du[2] = b * u[1] * u[2] - k * u[2]
        du[3] = k * u[2]
    end
    p0=[b,k]
    u0 = [1.0,i0,0.0]  
    tspan0 = (0.0,tend)
    prob = ODEProblem(covid!, u0, tspan0, p0)    
    sol=solve(prob)
    plot(sol, plotdensity=10000, vars=(1), label="Susceptible", title="result of epidemic simulation", xaxis="day", yaxis="fraction of population")
    plot!(sol, plotdensity=10000, vars=(2), label="Infected")
    plot!(sol, plotdensity=10000, vars=(3), label="Removed")
end
```
