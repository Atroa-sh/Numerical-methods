```julia
using Roots
using ForwardDiff
using DataFrames
#dla fractali newtona
using Fatou
using Makie
```

```julia
f1(x) = cos( x ) - x
f2(x) = x^3
f3(x) = x^3 - 2*x - 5
f4(x) = ( x + 3 ) * ( x - 1 )^2
f5(x) = exp( x ) - 1 / ( 10 * x )^2
f6(x) = 2 * x - exp( - x )
```

```julia
plot(f1)
plot!(f2)
plot!(f3)
plot!(f4)

```

```julia
x = find_zero(f3, (0, 3), Bisection(),verbose=true)
```

```julia
D(f3) = x->ForwardDiff.derivative(f3, float(x))
xdiff = find_zero((f3, D(f3)),0, Roots.Newton(),verbose=true)
sign(f3(prevfloat(xdiff))) *sign(f3(nextfloat(xdiff)))
```

```julia
find_zero(f3, 0, Order1(),  verbose=true)
#Roots.secant_method(f3, 0)
```

```julia
x1 = find_zero(f6, (-1, 1), Bisection(),verbose=true)
sign(f6(prevfloat(x1))) *sign(f6(nextfloat(x1)))
```

```julia
D(f6) = x->ForwardDiff.derivative(f6, float(x))
x2 = find_zero((f6,D(f6)),0, Roots.Newton(),verbose=true)
sign(f6(prevfloat(x2))) *sign(f6(nextfloat(x2)))
```

```julia
x3 = find_zero(f6, 1, Order1(),  verbose=true)
sign(f6(prevfloat(x3))) *sign(f6(nextfloat(x3)))
```

```julia
test(x) = x*exp(-x^2)
De(test) = x->ForwardDiff.derivative(test, float(x))
xt = find_zero((test,De(test)),1, Roots.Newton(),verbose=true)
sign(test(prevfloat(xt))) *sign(test(nextfloat(xt)))
#jesli zaczniemy w zlym punckie to nigdy nie dojdziemy
```

```julia
nf = newton(:(z^3-1),n=800,ϵ=0.1,N=25,iter=true,cmap="gist_earth", m=1)
Makie.plot(fatou(nf), bare=true)
```
